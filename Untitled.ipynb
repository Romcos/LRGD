```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import trange
```

```python
U = [1]
T = 10000000
for _ in trange(T):
    U.append(U[-1] + 4*U[-1]**(1/2))
```

```python
plt.plot(range(0,T+1,100),U[::100])
plt.plot(range(1,T,100),[n**2.1 for n in range(1,T,100)], color="red")
```
